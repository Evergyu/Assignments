```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import math

list_K=[2,5,10]
list_src=['1.jpg','2.jpg']
for K in list_K:
    for src in list_src:
        if(src=="1.jpg"):
            name="tiger"
        else:
            name="airplane"
        
        src = img.imread(src)
        
        #  * , 3(Dimension)    (  )
        feature_space = np.double(np.reshape(src, (-1, src.shape[2])))
        
        # feature_space.shape[0]      K  
        centroid_idx = np.random.choice(feature_space.shape[0], K, replace=False)
        
        #
        max_iter=50
        
        #3  k (kmeans)
        centroids = feature_space[centroid_idx]
        for i in range(max_iter): 
            #   
            for i in range(K):
            #    
                dist = [[math.dist(a, b) for b in centroids] for a in feature_space]
            
                #    
                idx = np.argmin(dist, axis=1)

            # Update centroids
            for k in range(K):
                centroids[k] = np.mean(feature_space[idx == k], axis=0)
            
        #    
        feature_space_temp=feature_space
        idx_temp=idx
        result = np.reshape(idx, (src.shape[0], src.shape[1]))
        
        # Display segmented image
        plt.imshow(result)
        plt.savefig("./"+name+"_"+str(K)+"_Cluster.png")
        plt.axis("off")
        print("done")
        plt.show()
```
