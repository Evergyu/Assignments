```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import math

list_K=[2,5,10]
list_src=['1.jpg','2.jpg']
for K in list_K:
    for src in list_src:
        if(src=="1.jpg"):
            name="tiger"
        else:
            name="airplane"
        
        src = img.imread(src)
        
        # 픽셀높이 * 넓이, 3(Dimension) 으로 차원 변환 (클러스터링 하기 위해)
        feature_space = np.double(np.reshape(src, (-1, src.shape[2])))
        
        # feature_space.shape[0] 중에서 초기 중심이 될 인덱스 K개 랜덤뽑기 
        centroid_idx = np.random.choice(feature_space.shape[0], K, replace=False)
        
        #반복수
        max_iter=50
        
        #3차원 점 k개 (kmeans)
        centroids = feature_space[centroid_idx]
        for i in range(max_iter): 
            #중심점 비교를 위한 변수
            for i in range(K):
            #중심점과 특징공간의 거리 측정 라벨수만큼
                dist = [[math.dist(a, b) for b in centroids] for a in feature_space]
            
                # 거리가 가장 작은걸로 라벨링
                idx = np.argmin(dist, axis=1)

            # Update centroids
            for k in range(K):
                centroids[k] = np.mean(feature_space[idx == k], axis=0)
            
        # 라벨링정보를 원래 이미지로 복원
        feature_space_temp=feature_space
        idx_temp=idx
        result = np.reshape(idx, (src.shape[0], src.shape[1]))
        
        # Display segmented image
        plt.imshow(result)
        plt.savefig("./"+name+"_"+str(K)+"_Cluster.png")
        plt.axis("off")
        print("done")
        plt.show()
```
